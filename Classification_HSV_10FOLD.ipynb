```python
def convertRowToFloat(string):
    x = string.split(",")
    result=[]
    for r in x:
        result.append(float(r))
    return result
```

```python
def createFeatureMatrix():
    results = []
    with open(path+feature_file_name) as csvfile:
        for row in csvfile:
            row_ = row[13:]
            results.append(convertRowToFloat(row_))
    return results
```

```python
def createLabelsVector(string):
    x_ = string.strip("\n")
    x = x_.split(";")
    result=[]
    for r in x:
        result.append(r)    
    return result 
```

```python
def getDataLabels():
    result=[]
    print(path+label_file_name)
    with open(path+label_file_name) as csvfile:
        for row in csvfile:
            row_=row[9:]
            result.append(createLabelsVector(row_))
    return result
```

```python
def DT(i):
    depth = [] 
    for j in range(1,101):
        depth.append(int(j))
    depth.append(None)
    parameter = [{
        "criterion":['gini','entropy'],
        "splitter":["best","random"],
        "max_depth":depth,
        "min_samples_split":[2,3,4,5,6,7,8,9,10]
    }]
    X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=i,random_state=0)
    clf = GridSearchCV(DecisionTreeClassifier(),param_grid=parameter,cv = 10)
    clf.fit(X_train,Y_train)
    porcentaje = i*100
    print("Se entrenó Arboles de Decisión con ",porcentaje,"% para pruebas")
    return clf
```

```python
def KNN(i):
    neighbors = [] 
    for j in range(1,101):
        neighbors.append(int(j))
    parameter = [{
        'n_neighbors':neighbors,
        'algorithm':['auto'],
        'weights':['uniform','distance'],
        'n_jobs':[-1]
    }]
    X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=i,random_state=0)
    clf = GridSearchCV(KNeighborsClassifier(),param_grid=parameter,cv = 10)
    clf.fit(X_train,Y_train)
    porcentaje = i*100
    print("Se entrenó K-Vecinos más cercanos con ",porcentaje,"% para pruebas")
    return clf
```
