```python
import csv
import numpy as np
from sklearn import svm
path='C:\\Users\\Cristhian\\Desktop\\'
path2='C:\\Users\\Cristhian\\Documents\\TEETH-RECOGNITION-WITH-MACHINE-LEARNING\\Labels\\'
feature_file_name = 'features.csv'
label_file_name='Labels.csv'
```

```python
def convertRowToFloat(string):
    x = string.split(",")
    result=[]
    for r in x:
        result.append(float(r))
    return result
```

```python
def createFeatureMatrix():
    results = []
    with open(path+feature_file_name) as csvfile:
        for row in csvfile:
            row_ = row[13:]
            results.append(convertRowToFloat(row_))
    return results
```

```python
def createLabelsVector(string):
    x = string.split(",")
    x = x.strip("\n")
    result=[]
    for r in x:
        result.append(r)
    print(result)    
    return result 
```

```python
def getDataLabels():
    result=[]
    with open(path2+label_file_name) as csvfile:
        for row in csvfile:
            row_=row[9:]
            result.append(createLabelsVector(row_))
        print(result)
```

```python
getDataLabels()
createFeatureMatrix()
```
